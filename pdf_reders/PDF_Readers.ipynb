## Save Images

```python
import os
import fitz  # pip install --upgrade pip; pip install --upgrade pymupdf
from tqdm import tqdm # pip install tqdm

workdir = "../../../pdf_files/"

for each_path in os.listdir(workdir):
    if ".pdf" in each_path:
        doc = fitz.Document((os.path.join(workdir, each_path)))

        for i in tqdm(range(len(doc)), desc="pages"):
            for img in tqdm(doc.get_page_images(i), desc="page_images"):
                xref = img[0]
                image = doc.extract_image(xref)
                pix = fitz.Pixmap(doc, xref)
                pix.save(os.path.join(workdir, "%s_p%s-%s.png" % (each_path[:-4], i, xref)))
                
print("Done!")
```

```python
Bold_words = []
Bold = ''

page = page_0
for i in range(len(page.extract_words(extra_attrs=["fontname", "size"]))):
    fontname = page.extract_words(extra_attrs=["fontname", "size"])[i]['fontname']
    size = page.extract_words(extra_attrs=["fontname", "size"])[i]['size']
    #sizes.append(size)
    Bold_word = {}
    if 'Bold' in fontname:
        Bold_word['fontname'] = fontname
        Bold_word['size'] = size
        word = page.extract_words(extra_attrs=["fontname", "size"])[i]['text']
        Bold_word['word'] = word
        Bold = Bold + word +' '
        
        Bold_words.append(Bold_word)

Bold_words
```

```python
import pandas as pd
import sys
import re

# for PDF to text conversion
from tika import parser
import pdf2image
import pytesseract
from PIL import Image

import pdfminer  # Version of pdfminer used: 20191125
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfpage import PDFPage
import io
import os
# detect language
from langdetect import detect
# os.environ['TIKA_SERVER_JAR'] = 'https://repo1.maven.org/maven2/org/apache/tika/tika-server/1.19/tika-server-1.19.jar'

##############################################################################
# Process PDF Functions
##############################################################################


def ocr_core(image):
    """
    This function will handle the core OCR processing of images.
    """
    # We'll use Pillow's Image class to open the image and pytesseract to detect the string in the image
    text = pytesseract.image_to_string(image)
    return text


def pdftopil(path_to_pdf):
    """
    This method reads a pdf and converts it into a sequence of images
    Input:
      path_to_pdf - as string, the path to the PDF file
    output:
      list of images
    """
    output_folder = '/'.join(path_to_pdf.split('/')[:-1])
  
    pil_images = pdf2image.convert_from_path(path_to_pdf,
                                           output_folder=output_folder,
                                           fmt='jpg',)
    return pil_images


def pdf2img2text(path_to_pdf):
    """
    This function extracts text from a pdf file using converting it to image and using OCR
    Input:
        path_to_pdf - as string, the path to the PDF file
        first_page  - as integer or None, a number of the first page to be processed by pdftoppm
        last_pageas - as integer or None, a number of the last page to be processed by pdftoppm
    output:
        extracted text
    """

    images = pdftopil(path_to_pdf=path_to_pdf)

    texts = []
    for image in images:
        text = ocr_core(image)
        texts.append(text)

    return ' '.join(texts)


def pdfminer_pdf_to_text(path):
    """
    Reads PDF file located at path into python text (string ready for processing).
    Input
        path - as string, that represent full path (folder+name) to the PDF file
    Output
        text as string.
    """
    rsrcmgr = PDFResourceManager()
    retstr = io.StringIO()
    laparams = LAParams()
    device = TextConverter(rsrcmgr, retstr, laparams=laparams)

    fp = open(path, 'rb')
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    password = ""
    maxpages = 0
    caching = True
    pagenos = set()

    for page in PDFPage.get_pages(fp, pagenos, maxpages=maxpages,
                                  password=password,
                                  caching=caching,
                                  check_extractable=True):
        interpreter.process_page(page)

    text = retstr.getvalue()

    fp.close()
    device.close()
    retstr.close()

    return text


def get_text_from_pdf(path_to_pdf):
    '''
    Extracts textual content from PDF file
    using any approach nessesary
    input:
        path_to_pdf - as string, the path to the PDF file
    output:
        extracted text
    '''
    text = ''
    if len(path_to_pdf) > 0:
        try:
            pdf_contents = parser.from_file(path_to_pdf)
            text = pdf_contents['content']
            # check the length of extracted text and do pdfminer_pdf_to_text if too short
            if len(str(text)) < 500:
                text = pdfminer_pdf_to_text(path_to_pdf)
            # check the length of extracted text and do OCR if too short
            if len(str(text)) < 500:
                text = pdf2img2text(path_to_pdf=path_to_pdf)
        except:
            return text
    return text
get_text_from_pdf("EIB_Group_sustainability_report_2020_en.pdf")
```
